```python
#Processing countries.scv file

data = pd.read_csv('countries.csv')

data
```

```python
country = data['country']
country
```

```python
pd.unique(country)
```

```python
# plotting between Afghanistan', 'Albania

afg = data[data.country == 'Afghanistan']
afg
```

```python
albn = data[data.country == 'Albania']
albn
```

```python
plt.plot(afg.year,afg.population/10**6)
plt.plot(albn.year,albn.population/10**6)

plt.xlabel('Year')
plt.ylabel('Population(in million)')
plt.title('Polpulation Comparison')

plt.legend(['Afghanistan population','Albania population'])

plt.show()
```
